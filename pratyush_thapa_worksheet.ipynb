Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
import pandas as pd


# Calculate the IQR for 'Score'
Q1_score = southasia['score'].quantile(0.25)
Q3_score = southasia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define lower and upper bounds for 'Score'
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outlier countries based on 'Score'
outliers_score = southasia[(southasia['score'] < lower_bound_score) | (southasia['score'] > upper_bound_score)]

# Calculate the IQR for 'GDP per Capita'
Q1_gdp = southasia['Log GDP per capita'].quantile(0.25)
Q3_gdp = southasia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define lower and upper bounds for 'GDP per Capita'
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outlier countries based on 'GDP per Capita'
outliers_gdp = southasia[(southasia['Log GDP per capita'] < lower_bound_gdp) | (southasia['Log GDP per capita'] > upper_bound_gdp)]

# Combine outliers based on 'Score' and 'GDP per Capita'
outliers_combined = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

# Display outlier countries
print("Outlier countries based on Score and GDP per Capita:")
# Change 'country name' to 'Country name' to match the actual column name
print(outliers_combined[['Country name', 'score', 'Log GDP per capita']])
```

```python
import pandas as pd


# Calculate the IQR for 'Score'
Q1_score = southasia['score'].quantile(0.25)
Q3_score = southasia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define lower and upper bounds for 'Score' using the 1.5 × IQR rule
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outlier countries based on 'Score'
outliers_score = southasia[(southasia['score'] < lower_bound_score) | (southasia['score'] > upper_bound_score)]

# Calculate the IQR for 'GDP per Capita'
Q1_gdp = southasia['Log GDP per capita'].quantile(0.25)
Q3_gdp = southasia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define lower and upper bounds for 'GDP per Capita' using the 1.5 × IQR rule
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outlier countries based on 'GDP per Capita'
outliers_gdp = southasia[(southasia['Log GDP per capita'] < lower_bound_gdp) | (southasia['Log GDP per capita'] > upper_bound_gdp)]

# Combine outliers based on 'Score' and 'GDP per Capita'
outliers_combined = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

# Display outlier countries
print("Outlier countries based on Score and GDP per Capita:")
print(outliers_combined[['Country name', 'score', 'Log GDP per capita']])
```

Task - 4 - Exploring Trends Across Metrics:
Tasks:
1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.


```python
from scipy import stats

# to filter data for south asia
southasia_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Sri Lanka', 'Pakistan', 'Nepal']
southasia_df = data[data['Country name'].isin(southasia_countries)]

# Metrics of interest
metrics = ['Freedom to make life choices', 'Generosity']

# Checking  if 'score' exists in the dataframe
if 'score' not in southasia_df.columns:
    print("Error: 'score' column is missing in the dataframe.")
else:
    # Calculate correlation with the score
    correlations = {}
    for metric in metrics:
        if metric not in southasia_df.columns:
            print(f"Error: '{metric}' column is missing in the dataframe.")
        else:
            correlation, _ = stats.pearsonr(southasia_df['score'], southasia_df[metric])
            correlations[metric] = correlation

    # Print the correlation values
    print("Pearson Correlation with Happiness Score for South Asian Countries:")
    print(correlations)

    # Create scatter plots with trendlines
    plt.figure(figsize=(12, 6))

    # Plot for 'Freedom to make life choices'
    plt.subplot(1, 2, 1)
    plt.scatter(southasia_df['Freedom to make life choices'], southasia_df['score'], color='blue')
    plt.title('Freedom to Make Life Choices vs Happiness Score')
    plt.xlabel('Freedom to Make Life Choices')
    plt.ylabel('Happiness Score')
    # Fit a trendline
    slope, intercept, _, _, _ = stats.linregress(southasia_df['Freedom to make life choices'], southasia_df['score'])
    plt.plot(southasia_df['Freedom to make life choices'], intercept + slope * southasia_df['Freedom to make life choices'], color='black')
    plt.grid(True)

    # Plot for 'Generosity'
    plt.subplot(1, 2, 2)
    plt.scatter(southasia_df['Generosity'], southasia_df['score'], color='red')
    plt.title('Generosity vs Happiness Score')
    plt.xlabel('Generosity')
    plt.ylabel('Happiness Score')
    # Fit a trendline
    slope, intercept, _, _, _ = stats.linregress(southasia_df['Generosity'], southasia_df['score'])
    plt.plot(southasia_df['Generosity'], intercept + slope * southasia_df['Generosity'], color='black')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Discussion of the strongest and weakest relationships
    print("Discussion:")
    for metric in metrics:
        correlation = correlations.get(metric, None)
        if correlation is not None:
            if correlation > 0:
                strength = "strong" if correlation > 0.5 else "moderate"
            else:
                strength = "weak"
            print(f"The relationship between '{metric}' and Happiness Score is {strength} with a correlation value of {correlation:.2f}.")
```

 Outlier Detection:

```python
# identify outliers based on 1.5*IQR rule
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# identify outliers in both regions
south_asia_outliers = find_outliers(south_asia_df, 'score')
middle_east_outliers = find_outliers(middle_east_df, 'score')

# plot outliers
plt.figure(figsize=(12, 6))
plt.scatter(southasia_df['Log GDP per capita'], southasia_df['score'], color='blue', label='South Asia')
plt.scatter(middle_east_df['Log GDP per capita'], middle_east_df['score'], color='orange', label='Middle East')
plt.scatter(south_asia_outliers['Log GDP per capita'], south_asia_outliers['score'], color='yellow', label='South Asia Outliers')
plt.scatter(middle_east_outliers['Log GDP per capita'], middle_east_outliers['score'], color='purple', label='Middle East Outliers')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Outliers in South Asia and Middle East')
plt.legend()
plt.show()
```
